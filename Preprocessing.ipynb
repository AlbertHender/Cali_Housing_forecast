```python
def get_weekly_dates(start_date_str, end_date_str=None):
    """
    Generate dates every week starting from the provided start date up to the end date.
    
    Args:
    - start_date_str: A string representing the start date in the format 'YYYY-MM-DD'.
    - end_date_str: A string representing the end date in the format 'YYYY-MM-DD'. If None, the current date is considered.
    
    Returns:
    - A list of datetime objects representing the weekly dates.
    """
    start_date = datetime.strptime(start_date_str, '%Y-%m-%d')
    
    if end_date_str:
        end_date = datetime.strptime(end_date_str, '%Y-%m-%d')
    else:
        end_date = datetime.now()
    
    weekly_dates = []
    current_date = start_date
    new_CPI = pd.DataFrame()
    while current_date <= end_date:
        weekly_dates.append(current_date.date())
        current_date += timedelta(weeks=1)
    
    return weekly_dates

# Example usage:
start_date = '2008-02-02'
end_date = '2024-02-20'  # You can provide an end date or leave it as None to get dates up to the current date
weekly_dates = get_weekly_dates(start_date, end_date)

test = pd.DataFrame({'DATE': weekly_dates, "num" :range(0,len(weekly_dates))})
test2 = data[0]
test2['DATE'] = test2['DATE'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
test = test.set_index('DATE')
test2 = test2.set_index('DATE')
test3 = pd.concat([test2,test],axis=1)
del test3['num']
new_CPI = test3.ffill().dropna()
new_CPI
```
